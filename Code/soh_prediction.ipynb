## Model Training

```python
# ============================================================================
# PREPARE DATA FOR MODELING
# ============================================================================
print("\n" + "="*80)
print("PREPARE DATA FOR MODELING")
print("="*80)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Verify final dataset
print(f"Features: {df_final.columns.tolist()}")

# Separate features and target
X = df_final
y = df_labels['soh_cap']

print(f"\nFeature set shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Target variable statistics:")
print(y.describe())

# Check for any remaining NaN values
print(f"\nNaN values in X: {X.isnull().sum().sum()}")
print(f"NaN values in y: {y.isnull().sum()}")

# Split data into training and testing sets
print("\n" + "-"*80)
print("SPLITTING DATA")
print("-"*80)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=123
)

print(f"Training set size: {X_train.shape[0]} ({100*X_train.shape[0]/len(X):.1f}%)")
print(f"Testing set size: {X_test.shape[0]} ({100*X_test.shape[0]/len(X):.1f}%)")

# Optional: Create validation set
# X_train_split, X_val, y_train_split, y_val = train_test_split(
#     X_train, y_train, test_size=0.2, random_state=42
# )

# print(f"Training set (after split): {X_train_split.shape[0]}")
# print(f"Validation set: {X_val.shape[0]}")

```

```python
# ============================================================================
# TRAIN ADVANCED TREE-BASED MODELS
# ============================================================================
print("\n" + "="*80)
print("TRAIN ADVANCED TREE-BASED REGRESSION MODELS")
print("="*80)

from sklearn.ensemble import (
    RandomForestRegressor, 
    GradientBoostingRegressor, 
    AdaBoostRegressor,
    ExtraTreesRegressor,
    HistGradientBoostingRegressor
)
from sklearn.tree import DecisionTreeRegressor
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor
import time

# Dictionary to store all models and results
models_dict = {}
results_dict = {}
```

### Train models

```python
# ============================================================================
# MODEL 1: RANDOM FOREST
# ============================================================================
print("\n" + "="*80)
print("MODEL 1: RANDOM FOREST REGRESSOR")
print("="*80)

print("\nTraining Random Forest...")
start_time = time.time()

rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=20,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1,
    verbose=1
)

rf_model.fit(X_train, y_train)
rf_train_time = time.time() - start_time

# Make predictions
y_train_pred_rf = rf_model.predict(X_train)
# y_val_pred_rf = rf_model.predict(X_val)
y_test_pred_rf = rf_model.predict(X_test)

# Calculate metrics
rf_train_r2 = r2_score(y_train, y_train_pred_rf)
# rf_val_r2 = r2_score(y_val, y_val_pred_rf)
rf_test_r2 = r2_score(y_test, y_test_pred_rf)

rf_train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred_rf))
# rf_val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred_rf))
rf_test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred_rf))

rf_train_mae = mean_absolute_error(y_train, y_train_pred_rf)
# rf_val_mae = mean_absolute_error(y_val, y_val_pred_rf)
rf_test_mae = mean_absolute_error(y_test, y_test_pred_rf)

print(f"\n✓ Random Forest Training Complete (Time: {rf_train_time:.2f}s)")
print(f"\nTraining Metrics:")
print(f"  R² Score: {rf_train_r2:.6f}")
print(f"  RMSE: {rf_train_rmse:.6f}")
print(f"  MAE: {rf_train_mae:.6f}")
# print(f"\nValidation Metrics:")
# print(f"  R² Score: {rf_val_r2:.6f}")
# print(f"  RMSE: {rf_val_rmse:.6f}")
# print(f"  MAE: {rf_val_mae:.6f}")
print(f"\nTest Metrics:")
print(f"  R² Score: {rf_test_r2:.6f}")
print(f"  RMSE: {rf_test_rmse:.6f}")
print(f"  MAE: {rf_test_mae:.6f}")

models_dict['Random Forest'] = rf_model
results_dict['Random Forest'] = {
    'train_r2': rf_train_r2, 'test_r2': rf_test_r2,
    'train_rmse': rf_train_rmse, 'test_rmse': rf_test_rmse,
    'train_mae': rf_train_mae, 'test_mae': rf_test_mae,
    'train_time': rf_train_time,
    'y_test_pred': y_test_pred_rf
}
# results_dict['Random Forest'] = {
#     'train_r2': rf_train_r2, 'val_r2': rf_val_r2, 'test_r2': rf_test_r2,
#     'train_rmse': rf_train_rmse, 'val_rmse': rf_val_rmse, 'test_rmse': rf_test_rmse,
#     'train_mae': rf_train_mae, 'val_mae': rf_val_mae, 'test_mae': rf_test_mae,
#     'train_time': rf_train_time,
#     'y_test_pred': y_test_pred_rf
# }
```

```python
# ============================================================================
# MODEL 2: GRADIENT BOOSTING
# ============================================================================
print("\n" + "="*80)
print("MODEL 2: GRADIENT BOOSTING REGRESSOR")
print("="*80)

print("\nTraining Gradient Boosting...")
start_time = time.time()

gb_model = GradientBoostingRegressor(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=7,
    min_samples_split=5,
    min_samples_leaf=2,
    subsample=0.8,
    random_state=42,
    verbose=0
)

gb_model.fit(X_train, y_train)
gb_train_time = time.time() - start_time

# Make predictions
y_train_pred_gb = gb_model.predict(X_train)
# y_val_pred_gb = gb_model.predict(X_val)
y_test_pred_gb = gb_model.predict(X_test)

# Calculate metrics
gb_train_r2 = r2_score(y_train, y_train_pred_gb)
# gb_val_r2 = r2_score(y_val, y_val_pred_gb)
gb_test_r2 = r2_score(y_test, y_test_pred_gb)

gb_train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred_gb))
# gb_val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred_gb))
gb_test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred_gb))

gb_train_mae = mean_absolute_error(y_train, y_train_pred_gb)
# gb_val_mae = mean_absolute_error(y_val, y_val_pred_gb)
gb_test_mae = mean_absolute_error(y_test, y_test_pred_gb)

print(f"\n✓ Gradient Boosting Training Complete (Time: {gb_train_time:.2f}s)")
print(f"\nTraining Metrics:")
print(f"  R² Score: {gb_train_r2:.6f}")
print(f"  RMSE: {gb_train_rmse:.6f}")
print(f"  MAE: {gb_train_mae:.6f}")
# print(f"\nValidation Metrics:")
# print(f"  R² Score: {gb_val_r2:.6f}")
# print(f"  RMSE: {gb_val_rmse:.6f}")
# print(f"  MAE: {gb_val_mae:.6f}")
print(f"\nTest Metrics:")
print(f"  R² Score: {gb_test_r2:.6f}")
print(f"  RMSE: {gb_test_rmse:.6f}")
print(f"  MAE: {gb_test_mae:.6f}")

models_dict['Gradient Boosting'] = gb_model
results_dict['Gradient Boosting'] = {
    'train_r2': gb_train_r2, 'test_r2': gb_test_r2,
    'train_rmse': gb_train_rmse, 'test_rmse': gb_test_rmse,
    'train_mae': gb_train_mae, 'test_mae': gb_test_mae,
    'train_time': gb_train_time,
    'y_test_pred': y_test_pred_gb
}
# results_dict['Gradient Boosting'] = {
#     'train_r2': gb_train_r2, 'val_r2': gb_val_r2, 'test_r2': gb_test_r2,
#     'train_rmse': gb_train_rmse, 'val_rmse': gb_val_rmse, 'test_rmse': gb_test_rmse,
#     'train_mae': gb_train_mae, 'val_mae': gb_val_mae, 'test_mae': gb_test_mae,
#     'train_time': gb_train_time,
#     'y_test_pred': y_test_pred_gb
# }
```

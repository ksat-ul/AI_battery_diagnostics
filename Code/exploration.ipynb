```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
pd.set_option('display.max_columns', None)
```

```python
# Load the Excel file
df = pd.read_csv('../Data/cell_eocv2/cell_eocv2_P003_2_S02_C11.csv',sep=';')


df.head()
```

```python
df.describe()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['total_q_chg_other_RT_Ah'], autolag='AIC')

# 3. Format and display results
output = pd.Series(result[0:4], 
                   index=['Test Statistic', 'p-value', 'Lags Used', 'Number of Observations'])

for key, value in result[4].items():
    output[f'Critical Value ({key})'] = value

print("--- ADF Test Results ---")
print(output)

# 4. Interpret the p-value
if output['p-value'] <= 0.05:
    print("\nResult: The series is stationary (Reject H0)")
else:
    print("\nResult: The series is non-stationary (Fail to reject H0)")
```

```python
corr_matrix = df[['delta_q_Ah', 'delta_q_chg_Ah',
       'delta_q_dischg_Ah', 'delta_e_Wh', 'delta_e_chg_Wh',
       'delta_e_dischg_Wh',
       'ocv_est_start_V', 'ocv_est_end_V', 'soc_est_start', 'soc_est_end',
       'num_cycles_checkup', 'total_q_chg_CU_RT_Ah', 'total_q_dischg_CU_RT_Ah',
       'total_q_chg_cyc_OT_Ah', 'total_q_dischg_cyc_OT_Ah',
       'total_q_chg_other_RT_Ah', 'total_q_dischg_other_RT_Ah',
       'total_q_chg_other_OT_Ah', 'total_q_dischg_other_OT_Ah',
       'total_q_chg_sum_Ah', 'total_q_dischg_sum_Ah', 'total_e_chg_CU_RT_Wh',
       'total_e_dischg_CU_RT_Wh', 'total_e_chg_cyc_OT_Wh',
       'total_e_dischg_cyc_OT_Wh', 'total_e_chg_other_RT_Wh',
       'total_e_dischg_other_RT_Wh', 'total_e_chg_other_OT_Wh',
       'total_e_dischg_other_OT_Wh', 'total_e_chg_sum_Wh',
       'total_e_dischg_sum_Wh']].corr(numeric_only=True)

print("Correlation Matrix:")
#print(corr_matrix)

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# 3. Visualize using a Heatmap (Seaborn)
plt.figure(figsize=(24, 24))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```
